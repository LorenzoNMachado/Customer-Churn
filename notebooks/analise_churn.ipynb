```python
import pandas as pd
import pandas as pd
import numpy as np
import math
from IPython.display import display
df = pd.read_csv('../data/WA_Fn-UseC_-Telco-Customer-Churn.csv')
def criar_tabela_frequencia(df, coluna):
    dados = df[coluna].dropna()
    n = len(dados)
    k = math.ceil(1 + 3.322 * math.log10(n))
    AT = dados.max() - dados.min()
    h = AT / k
    bins = [dados.min() + i * h for i in range(k + 1)]
    classes = pd.cut(dados, bins=bins, right=False, include_lowest=True)
    fi = classes.value_counts().sort_index() 
    
    tabela = pd.DataFrame({'Frequência Absoluta (fi)': fi})
    tabela.insert(0, 'Ponto Médio (xi)', [f"{(b.left + b.right) / 2:.2f}" for b in tabela.index])
    tabela['Frequência Relativa (fr)'] = tabela['Frequência Absoluta (fi)'] / n
    tabela['Fr (%)'] = tabela['Frequência Relativa (fr)'] * 100
    tabela['Freq. Acumulada (Fi)'] = tabela['Frequência Absoluta (fi)'].cumsum()
    tabela['Fr (%)'] = tabela['Fr (%)'].apply(lambda x: f"{x:.2f}%")
    tabela.index.name = f'Classes ({coluna})'
    tabela.index = tabela.index.map(lambda x: f"{x.left:,.2f} |--- {x.right:,.2f}")
    
    return tabela
```

```python
# PAINEL DE VARIÁVEIS DISPONÍVEIS
print("LISTA DE VARIÁVEIS PARA ANÁLISE")
print("-" * 55)
colunas_numericas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
colunas_categoricas = df.select_dtypes(include=['object']).columns.tolist()

print(" VARIÁVEIS NUMÉRICAS")
print("   (Use estas para a Tabela com Intervalos de Classe):")
for col in colunas_numericas:
    print(f"   -> {col}")

print("\n VARIÁVEIS CATEGÓRICAS / TEXTO ---- Não tem ocmo usar variaveis não numéricas ainda ")
print("   (Use estas para tabelas de frequência simples):")
for col in colunas_categoricas:
    if col != 'customerID':
        print(f"   -> {col}")
```

```python
# INTERAÇÃO COM O USUÁRIO
print("Variáveis contínuas sugeridas para análise com intervalos:")
print("- tenure (Meses de contrato)")
print("- MonthlyCharges (Cobrança Mensal)")
print("- TotalCharges (Cobrança Total Acumulada)\n")
variavel_escolhida = input("Digite exatamente o nome da variável que deseja analisar: ")

if variavel_escolhida in df.columns:
    if pd.api.types.is_numeric_dtype(df[variavel_escolhida]):
        print(f"\nGerando Tabela de Distribuição de Frequência para: {variavel_escolhida}\n")
        tabela_final = criar_tabela_frequencia(df, variavel_escolhida)
        display(tabela_final)
    else:
        print("Erro: A variável escolhida não é numérica (contínua/discreta).")
else:
    print("Erro: Variável não encontrada. Verifique se digitou corretamente (diferencia maiúsculas e minúsculas).")
```
